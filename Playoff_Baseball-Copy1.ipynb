```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset
df = pd.read_csv('MLB_baseball.csv')
```

```python
#Delete the RankPlayoffs and RankSeason columns because they are not needed for this analysis
df = df.drop(columns = ['RankPlayoffs', 'RankSeason'])
```

```python
#Create a new column for Total Runs +/-
df['Runs +/-'] = df['RS'] - df['RA']
```

```python
#Create a new column for Total OBP +/-
df['OBP +/-'] = df['OBP'] - df['OOBP']
```

```python
#Create a new column for Total SLG +/-
df['SLG +/-'] = df['SLG'] - df['OSLG']
```

```python
#Fill in the NaN values with the average of that column
df['OOBP'].fillna(value = df['OOBP'].mean(), inplace=True)
df['OSLG'].fillna(value= df['OSLG'].mean(), inplace = True)
df['SLG'].fillna(value= df['SLG'].mean(), inplace = True)
```

```python
#Correlation between Making Playoffs, Wins, Batting Average, Slugging %, On-Base %, Slugging % +/-, and On-Base % +/-
df[['Playoffs', 'SLG +/-','OBP +/-', 'BA', 'W', 'SLG', 'OBP']].corr()
```

```python
#Correlation heatmap of previous chart
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(df[['W', 'RS','RA','BA','SLG','OBP','SLG +/-','OBP +/-']].corr(), cmap=cmap, vmax=1, center=0,
            square=True, linewidths=1, cbar_kws={"shrink":1})
```
